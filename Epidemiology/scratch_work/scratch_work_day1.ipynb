# num of people getting vaccinated

```python
df_new_vax = df_US[["date", "new_vaccinations"]]
```

```python
df_new_vax.sample(10)
```

```python
df_new_vax["date"] = pd.to_datetime(df_new_vax["date"])
df_new_vax = df_new_vax.sort_values(by="date")

# Define the date range you want to plot
start_date1 = pd.to_datetime('2021-03-01')
end_date1 = pd.to_datetime('2021-06-01')

# Filter the DataFrame to include only the data within the specified date range
df_new_vax_mar_june = df_new_vax.loc[(df_new_vax["date"] >= start_date1) & (df_new_vax["date"] <= end_date1)]
```

```python
plt.figure(figsize=(18, 6))
plt.plot(df_new_vax_mar_june["date"], df_new_vax_mar_june["new_vaccinations"], linestyle='-', color='r')
plt.xlabel("Date")
plt.ylabel("Number of Vaccines Administered (per 100,000)")
plt.title('Number of People Receving Vaccines in the US from Mar-Jun 2021')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

def hunds_thousands_formatter(x, pos):
    return f'{x/100000:.0f}'

# Apply the custom formatter to the y-axis
plt.gca().yaxis.set_major_formatter((hunds_thousands_formatter))
plt.grid(True)
plt.savefig("number_of_new_people_getting_vax_mar-jun_2021.png")
plt.show()
```
