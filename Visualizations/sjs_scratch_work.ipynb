```python
tx_data = pd.read_csv("texas-history.csv")
tx_df = pd.DataFrame(tx_data)
```

```python
tx_df
```

```python
plot_tx = tx_df[["date","hospitalizedCurrently"]]
```

```python
plot_tx = tx_df.sort_values(by="date")
plot_tx
```

```python
x_values = plot_tx["date"]
```

```python
y_values = plot_tx["hospitalizedCurrently"]
```

```python
plt.figure(figsize=(14,10))
plt.plot(x_values, y_values, color="r", linewidth=3)

ticks = plt.xticks(plot_tx["date"].values[::30], rotation=25)

plt.title("Number of Texan's Hospitalized due to COVID-19", fontsize=18)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Number Hospitalized", fontsize=14)

plt.grid()
plt.savefig("texas_hospitalizations.png")
plt.show()
```

```python
hospital_df = tx_df[["date", "positive", "hospitalizedCurrently", "onVentilatorCurrently",  "inIcuCurrently"]].sort_values(by="date")
```

```python
hospital_df
```

```python

x1 = hospital_df["date"]
y1 = hospital_df["positive"]
y2 = hospital_df["hospitalizedCurrently"]
y3 = hospital_df["onVentilatorCurrently"]
y4 = hospital_df["inIcuCurrently"]
```

```python
plt.figure()

# positive
plt.subplot(221)
plt.plot(x1, y1)
plt.ylabel("Number of Positive Cases")
plt.xlabel("Date")
plt.title("Infected")
ticks = plt.xticks(hospital_df["date"].values[::30], rotation=25)
plt.grid(True)

# hospitalized
plt.subplot(222)
plt.plot(x1, y2)
#plt.yscale('log')
plt.title('log')
plt.xlabel("Date")
ticks = plt.xticks(hospital_df["date"].values[::30], rotation=25)
plt.grid(True)

# on ventilator
plt.subplot(223)
plt.plot(x1, y3)
#plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.xlabel("Date")
ticks = plt.xticks(hospital_df["date"].values[::30], rotation=25)
plt.grid(True)

# in icu
plt.subplot(224)
plt.plot(x1, y4)
#plt.yscale('logit')
plt.title('logit')
plt.xlabel("Date")
ticks = plt.xticks(hospital_df["date"].values[::30], rotation=25)
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
