```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
url = f"https://storage.googleapis.com/covid19-open-data/v3/location/US.csv"
usaData = pd.read_csv(url)
usaData.interpolate(inplace=True)
```

```python
ruralColumns = ["date" ,"cumulative_confirmed", "cumulative_deceased", "population_rural"]
ruralDataFrame = usaData[ruralColumns].set_index("date")
urbanDataFrame = usaData[urbanColumns].set_index("date")
```

```python
bothDataFrame= ruralDataFrame.merge(urbanDataFrame)
```

```python
bothDataFrame
bothDataFrame.interpolate(inplace=True)
```

```python
bothDataFrame
```

```python
fig = px.bar(bothDataFrame, y=["cumulative_deceased","cumulative_confirmed", "population_rural", "population_urban"])
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Number of People")
fig.write_image("images/uralAndCovid.png")
fig.show()
```
